```python
# Dependecies
import pandas as pd
import numpy as np

FileToLoad = "Resources/purchase_data.csv"

PurData = pd.read_csv(FileToLoad,  encoding="ISO-8859-1")
```

```python
PurData.head()
```

## Player Count

* Display the total number of players


```python
#Player Count
CountPlayer = PurData["SN"].nunique()

PlayerList = [{"Total Players": CountPlayer}]

PlayerDf   = pd.DataFrame(PlayerList)

PlayerDf
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Purchasing Analysis (Total)
ItemCount = PurData["Item ID"].nunique()

AvePrice = round(PurData["Price"].mean(),2)

NumPur = PurData["Purchase ID"].nunique() 

TotalRev = round(PurData["Price"].sum(),2)

PurDetail = [{"Number of Unique Items": ItemCount, "Average Price": AvePrice ,"Number of Purchases": NumPur,"Total Revenue":TotalRev}]

PurSumm = pd.DataFrame(PurDetail)

OrgDf = PurSumm[["Number of Unique Items","Average Price","Number of Purchases","Total Revenue"]]

OrgDf["Average Price"] = OrgDf["Average Price"].map("${:.2f}".format)

OrgDf["Total Revenue"] = OrgDf["Total Revenue"].map("${:.2f}".format)

OrgDf
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# Gender Demographics
GenDetail = PurData.groupby("Gender").SN.nunique().sort_values(ascending=False).reset_index(name='Total Count')

Percent = round((GenDetail["Total Count"] / CountPlayer)*100,2)

GenDetail["Percent of Players"] = Percent

GenDetail = GenDetail.set_index("Gender")

GenDetail
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Purchasing Analysis (Gender)
GroupPur = PurData.groupby(["Gender"])

PriceGen = GroupPur["Price"].sum()

PurGen = pd.DataFrame(PriceGen)

PurGen = PurGen.rename(columns = {"Price":"Total Purchase Value"})

PurGen["Purchase count"] = PurData.groupby("Gender").SN.count()

PurGen["Average Purchase Price"] = round(PurGen["Total Purchase Value"] / PurGen["Purchase count"],2)

PurGen["Avg Total Purchase per Person"] = round(PurGen["Total Purchase Value"] / GenDetail["Total Count"],2)

PurGen = PurGen[["Purchase count","Average Purchase Price","Total Purchase Value","Avg Total Purchase per Person"]]

PurGen["Total Purchase Value"] = PurGen["Total Purchase Value"].map("${:.2f}".format)

PurGen["Average Purchase Price"] = PurGen["Average Purchase Price"].map("${:.2f}".format)

PurGen["Avg Total Purchase per Person"] = PurGen["Avg Total Purchase per Person"].map("${:.2f}".format)

PurGen
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Age Demographics
BinsAge = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]

BinNames = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

PurData["Age Group"] = pd.cut(PurData["Age"],BinsAge, labels=BinNames)

PurData

GrpAge = PurData.groupby("Age Group")

CountAge = GrpAge["SN"].nunique()

PercentAge = (CountAge/CountPlayer) * 100

DemoAge = pd.DataFrame({"Total Count": CountAge, "Percentage of Players": PercentAge})

DemoAge.index.name = None

DemoAge.style.format({"Percentage of Players":"{:,.2f}"})
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Purchasing Analysis (Age)
PurchasesAge = PurData.groupby(["Age Group"])

PriceAge = PurchasesAge["Price"].sum()

AgePur = pd.DataFrame(PriceAge)

AgePur = AgePur.rename(columns={"Price":"Total Purchase Value"})

AgeCount = PurchasesAge["Age Group"].count()

AgePur["Purchase Count"] = pd.DataFrame(AgeCount)

AgeCount = PurchasesAge["SN"].nunique()

AgePur["Total Count"]  = pd.DataFrame(AgeCount)

AgePur["Average Purchase Price"]=round(AgePur["Total Purchase Value"]/AgePur["Purchase Count"] ,2).sort_values(ascending=True)

AgePur["Avg Total Purchase per Person"]=round(AgePur["Total Purchase Value"]/AgePur["Total Count"],2).sort_values(ascending=True)

AgePur=AgePur[["Purchase Count","Average Purchase Price","Total Purchase Value","Avg Total Purchase per Person"]]

AgePur.index.name = None

AgePur["Average Purchase Price"] = AgePur["Average Purchase Price"].map("${:.2f}".format)

AgePur["Total Purchase Value"] = AgePur["Total Purchase Value"].map("${:.2f}".format)

AgePur["Avg Total Purchase per Person"] = AgePur["Avg Total Purchase per Person"].map("${:.2f}".format)

AgePur
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Top Spenders
SpendPur = PurData.groupby(["SN"])

SpendPrice = SpendPur["Price"].sum().sort_values(ascending=False)

Spender = pd.DataFrame(SpendPrice)

Spender = Spender.rename(columns = {"Price":"Total Purchase Value"})

Spender.sort_values(["Total Purchase Value"], ascending=False)

SpendCount = PurData.groupby("SN").count()

SpendDetail = pd.DataFrame(SpendCount)

Spender["Purchase Count"] = SpendDetail["Purchase ID"]

Spender["Average Purchase Price"] = round((Spender["Total Purchase Value"]/Spender["Purchase Count"]),2)

Spender=Spender[["Purchase Count","Average Purchase Price","Total Purchase Value"]]

Spender["Total Purchase Value"] = Spender["Total Purchase Value"].map("${:.2f}".format)

Spender["Average Purchase Price"] = Spender["Average Purchase Price"].map("${:.2f}".format)

Spender.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Most Popular Items
Itm = PurData.groupby(["Item ID", "Item Name"])

PurCount = Itm['Price'].count()

Price = Itm['Price'].mean()

TotalPrice = Itm["Price"].sum()

PopItmDF = pd.DataFrame({'Purchase Count': PurCount, "Item Price": Price, "Total Purchase Value": TotalPrice})

PopItmDF["Item Price"] = PopItmDF["Item Price"].map("${:.2f}".format)

PopItmDF.sort_values("Purchase Count", ascending=False).head()
```
